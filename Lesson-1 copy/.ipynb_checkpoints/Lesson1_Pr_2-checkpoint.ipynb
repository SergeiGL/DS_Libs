### Problem 1

```python
import pandas as pd
```

```python
authors = pd.DataFrame({
    "author_id":[1, 2, 3],
    "author_name":['Тургенев', 'Чехов', 'Островский']
})
```

```python
authors
```

```python
books = pd.DataFrame({
    "author_id": [1, 1, 1, 2, 2, 3, 3],
    "book_title": ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
    "price": [450, 300, 350, 500, 450, 370, 290]
})
```

```python
books
```

### Problem 2

```python
authors_price = pd.merge(authors, books, on = 'author_id', how = 'right')
```

```python
authors_price
```

### Problem 3

```python
top = authors_price.sort_values(by='price', ascending = False)
```

```python
top5= top[:5]
```

```python
top5
```

### Problem 4

```python
authors_ = authors_price.groupby("author_name")
```

```python
min_p = authors_["price"].min()
```

```python
max_p = authors_["price"].max()
```

```python
mean_p = authors_["price"].mean()
mean_p
```

```python
authors_price_merged = pd.merge(min_p, max_p,on = 'author_name', how = 'right')
authors_price_merged
```

```python
authors_stat = pd.merge(authors_price_merged,mean_p ,on = 'author_name', how = 'right')
authors_stat.rename(columns={'price_x': 'min_price','price_y': 'max_price', 'price': 'mean_price'},  inplace=True)
authors_stat
```
