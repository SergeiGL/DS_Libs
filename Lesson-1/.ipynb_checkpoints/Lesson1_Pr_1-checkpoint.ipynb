### Problem 1

```python
import numpy as np
```

```python
a = np.array([[1,6], 
             [2,8],
             [3,11],
             [3, 10],
             [1, 7]])
```

```python
a
```

```python
mean_a = a.mean(axis = 0)
```

```python
mean_a
```

### Problem 2

```python
a_centered = a - mean_a
```

```python
a_centered
```

### Problem 3

```python
a_centered_sp_n = a_centered[:,0]@a_centered[:,1]
```

```python
a_centered_sp_n
```

```python
a_centered_sp = a_centered_sp_n / (a_centered.shape[0]-1)
```

```python
a_centered_sp
```
