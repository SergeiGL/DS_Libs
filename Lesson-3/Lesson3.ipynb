### Problem 1

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
boston = load_boston()
data = boston["data"]
target = boston["target"]
feature_names = boston["feature_names"]
X = pd.DataFrame(data, columns=feature_names)
Y = pd.DataFrame(target, columns=["price"])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
Y_pred = lr.predict(X_test)
```

```python
check_test = pd.DataFrame({
    "y_test": Y_test["price"],
    "y_pred": Y_pred.flatten(),
})
lr_r2=r2_score(check_test.y_test,check_test.y_pred)
lr_r2
```

### Problem 2

```python
RFR = RandomForestRegressor(n_estimators=1000,max_depth=12,random_state=42)
RFR.fit(X_train, Y_train.values[:, 0])
Y_pred_rfr = RFR.predict(X_test)
```

```python
check_test_rfr = pd.DataFrame({
    "y_test": Y_test["price"],
    "y_pred": Y_pred_rfr.flatten(),
})
rfr_r2=r2_score(check_test_rfr.y_test,check_test_rfr.y_pred)

print(f"RandomForest: {rfr_r2}\nLinearRegression: {lr_r2}")
```

#### RandomForest is better for this case
